(1). Create Histogram for variable Temp (i.e. numeric Temperature, degrees F).

```r
hist(Temp, col="bisque", xlab="Temp(degree F)") 
#The temperature was maximum around 80-85 degree F for a period of 35 days.
```

(2). Create Histogram for variable Temp with 20 breaks (i.e. numeric Temperature, degrees F).

```r
hist(Temp, breaks = 20,col="azure3", xlab="Temp (degree F)", main="Histogram of Temp with 20 breaks")
# With 20 breaks histogram, we can see that temperature was maximum around 80-85 degree F for a period of 20 days.
```

(3). Plot BOXPLOT for variable Ozone (i.e. numeric Ozone, ppb).

```r
boxplot(Ozone,col="bisque",outcol="red",pch=20, main ="Box plot of Ozone", ylab="Ozone(ppb)")
# From the boxplot of Ozone, we can clearly see that there are two outliers. The data set has more observations in 0 – 60 ppb
```

(4). Plot SCATTER PLOT for variables (Temp, Ozone).

```r
plot(Temp,Ozone, col="violetred",pch=20, main="Ozone vs Temp", ylab="Ozone(ppb)", xlab="Temp(degree F)")
# From the above plot, we can see that there is a linear relationship between the variables Ozone and Temp
```
